```python
from scipy.interpolate import CubicSpline
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import numpy as np

def read_csv():
    with open("power_generation.csv") as csvfile:  # same location as data3.csv
        datareader = csv.reader(csvfile, delimiter=',')
        times = []
        categories = []
        num_categories = len(next(datareader)) - 1  # calculate number of columns
        csvfile.seek(0)  # reset dataset after the next(...) function call

        for i in range(num_categories):
            categories.append([])
        for row in datareader:
            times.append((row[0]))
            for i in range(num_categories):
                categories[i].append((row[i + 1]))

        returned_values = []
        returned_values.append(times)
        for i in range(num_categories):
            returned_values.append(categories[i])

    return returned_values

category_names = ["Timestamps", "Geothermal", "Biomass", "Biogas", "Small Hydro", "Wind Total",
                  "Solar PV", "Solar Thermal", "Renewable", "Nuclear", "Thermal", "Imports", "Hydro"]
values = read_csv()
orig_timestamps = values[0]
del orig_timestamps[0]
geothermal_values = values[1]
del geothermal_values[0]
biomass_values = values[2]
del biomass_values[0]
biogas_values = values[3]
del biogas_values[0]
small_hydro_values = values[4]
del small_hydro_values[0]
wind_total_values = values[5]
del wind_total_values[0]
solar_pv_values = values[6]
del solar_pv_values[0]
solar_thermal_values = values[7]
del solar_thermal_values[0]
renewable_values = values[8]
del renewable_values[0]
nuclear_values = values[9]
del nuclear_values[0]
thermal_values = values[10]
del thermal_values[0]
imports_values = values[11]
del imports_values[0]
hydro_values = values[12]
del hydro_values[0]

def make_cubic_spline(x, y):
    f = CubicSpline(x, y, bc_type='natural')
    interpolated_x_values = shifting_timestamps(x)
    x_new = np.linspace(interpolated_x_values[0], interpolated_x_values[len(interpolated_x_values) - 1], len(interpolated_x_values))
    y_new = f(x_new)
    return y_new

def convert_time(x):
    new_times = []
    for i in x:
        dt_example = i
        dt_str = dt_example.split(".")
        new_times.append(dt_str)
    single_int_list = []
    large_int_list = []
    for j in new_times:
        for k in j:
            single_int_list.append(int(k))
        large_int_list.append(single_int_list)
        single_int_list = []
    large_dt_list = []
    for l in large_int_list:
        single_date_time = datetime(l[0], l[1], l[2], l[3] - 1)
        large_dt_list.append(single_date_time)
    converted_list = []
    dt_baseline = datetime(2000, 1, 1)
    for m in large_dt_list:
        time_diff = m - dt_baseline
        total_hrs = time_diff.total_seconds()/3600
        converted_list.append(total_hrs)
    return converted_list

def making_new_timestamps(x):
    new_list = []
    for i in x:
        new_list.append(i)
        new_list.append(i + 0.5)
    return new_list

def shifting_timestamps(x):
    new_list = []
    for i in x:
        new_list.append(i + 0.5)
    return new_list

def concatenate_list(y, y_new):
    new_list = []
    for i in range(len(y)):
        new_list.append(str(y[i]))
        new_list.append(str(y_new[i]))
    return new_list

orig_converted_times = convert_time(orig_timestamps)
new_times_list = making_new_timestamps(orig_converted_times)

new_geo_values = make_cubic_spline(orig_converted_times, geothermal_values)
new_geo_list = concatenate_list(geothermal_values, new_geo_values)

new_biomass_values = make_cubic_spline(orig_converted_times, biomass_values)
new_biomass_list = concatenate_list(biomass_values, new_biomass_values)

new_biogas_values = make_cubic_spline(orig_converted_times, biogas_values)
new_biogas_list = concatenate_list(biogas_values, new_biogas_values)

new_small_hydro_values = make_cubic_spline(orig_converted_times, small_hydro_values)
new_small_hydro_list = concatenate_list(small_hydro_values, new_small_hydro_values)

new_wind_total_values = make_cubic_spline(orig_converted_times, wind_total_values)
new_wind_total_list = concatenate_list(wind_total_values, new_wind_total_values)

new_solar_pv_values = make_cubic_spline(orig_converted_times, solar_pv_values)
new_solar_pv_list = concatenate_list(solar_pv_values, new_solar_pv_values)

new_solar_thermal_values = make_cubic_spline(orig_converted_times, solar_thermal_values)
new_solar_thermal_list = concatenate_list(solar_thermal_values, new_solar_thermal_values)

new_renewable_values = make_cubic_spline(orig_converted_times, renewable_values)
new_renewable_list = concatenate_list(renewable_values, new_renewable_values)

new_nuclear_values = make_cubic_spline(orig_converted_times, nuclear_values)
new_nuclear_list = concatenate_list(nuclear_values, new_nuclear_values)

new_thermal_values = make_cubic_spline(orig_converted_times, thermal_values)
new_thermal_list = concatenate_list(thermal_values, new_thermal_values)

new_imports_values = make_cubic_spline(orig_converted_times, imports_values)
new_imports_list = concatenate_list(imports_values, new_imports_values)

new_hydro_values = make_cubic_spline(orig_converted_times, hydro_values)
new_hydro_list = concatenate_list(hydro_values, new_hydro_values)

def total_timestamps(x):
    new_list = []
    for i in x:
        new_list.append(i)
        new_list.append(i + ".30")
    return new_list

final_list = []
final_dates = total_timestamps(orig_timestamps)

final_dates.insert(0, category_names[0])
new_geo_list.insert(0, category_names[1])
new_biomass_list.insert(0, category_names[2])
new_biogas_list.insert(0, category_names[3])
new_small_hydro_list.insert(0, category_names[4])
new_wind_total_list.insert(0, category_names[5])
new_solar_pv_list.insert(0, category_names[6])
new_solar_thermal_list.insert(0, category_names[7])
new_renewable_list.insert(0, category_names[8])
new_nuclear_list.insert(0, category_names[9])
new_thermal_list.insert(0, category_names[10])
new_imports_list.insert(0, category_names[11])
new_hydro_list.insert(0, category_names[12])

final_list.append(final_dates)
final_list.append(new_geo_list)
final_list.append(new_biomass_list)
final_list.append(new_biogas_list)
final_list.append(new_small_hydro_list)
final_list.append(new_wind_total_list)
final_list.append(new_solar_pv_list)
final_list.append(new_solar_thermal_list)
final_list.append(new_renewable_list)
final_list.append(new_nuclear_list)
final_list.append(new_thermal_list)
final_list.append(new_imports_list)
final_list.append(new_hydro_list)

np.savetxt('interpolated_power_generation.csv', final_list, fmt="%s", delimiter=',')
```
